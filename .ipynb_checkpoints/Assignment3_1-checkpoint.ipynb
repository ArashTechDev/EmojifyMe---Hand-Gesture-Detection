```python
df = df.dropna()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
numeric_df = df.select_dtypes(include=['number'])

numeric_df
```

```python
# normalization
scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(numeric_df)
```

```python
# Create DataFrame with numeric column names
scaled_df = pd.DataFrame(normalized_data, columns=numeric_df.columns)
# add column class to the normalized dataset
scaled_df['class'] = df['class']
```

```python
scaled_df.head(10)
```

```python
scaled_df.describe().T
```

```python
scaled_df.describe().T.round(1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.model_selection import train_test_split


input_train, input_test, output_train, output_test = train_test_split(
    normalized_data, 
    df['class'], 
    test_size=0.2, 
    random_state=5
)
```

```python
output_test
```

```python
output_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNN = KNeighborsClassifier()
KNN
```

```python
Knn = KNN.fit(input_train,output_train)
Knn
```

```python
pred_test = Knn.predict(input_test)
pred_test
```

```python
output_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score_knn = round(accuracy_score(output_test,pred_test)*100,0)
accuracy_score_knn
```

```python
from sklearn.metrics import confusion_matrix, classification_report

```

```python
cm = confusion_matrix(output_test, pred_test)
print("Confusion Matrix:")
print(cm)
```

```python
report = classification_report(output_test, pred_test)
print("\nClassification Report:")
print(report)
```
